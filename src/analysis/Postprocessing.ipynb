```python
import pandas as pd

df  = pd.read_csv("/content/deepvar_forecast_40step.csv")
```

```python
df = df.drop(["pred_lo50", "pred_hi50", "pred_lo90", "pred_hi90", "time_idx", "step"], axis=1)
df.head()
```

```python
df = df.rename(columns={"pred_mean": "value"})
df.head()
```

```python
df_wide = df.pivot(
    index=['baranggay_id', 'year', 'quarter'],
    columns='land_cover_type',
    values='value'
).reset_index()
df_wide.head()
```

```python
print(df_wide.columns)
```

```python
df_wide.to_csv('inference.csv', index=False)
```

```python
df1 = pd.read_csv("/content/time-series-data.csv")
df2 = pd.read_csv("/content/inference.csv")

print(set(df1.columns) == set(df2.columns))
df = pd.concat([df1, df2], ignore_index=True)
df = df.sort_values(by=["baranggay_id", "year", "quarter"])
```

```python
mapping = pd.read_csv("/content/barangay_mapping.csv")

df1 = df1.merge(mapping, on="baranggay_id", how="left")
df2 = df2.merge(mapping, on="baranggay_id", how="left")

df = pd.concat([df1, df2], ignore_index=True)
df = df.sort_values(by=["baranggay_id", "year", "quarter"])
```

```python
df.to_csv('training_inference.csv', index=False)
```
